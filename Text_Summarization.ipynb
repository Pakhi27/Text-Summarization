```python

from transformers import pipeline, set_seed

import matplotlib.pyplot as plt
from datasets import load_dataset
import pandas as pd
from datasets import load_dataset, load_metric

from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

import nltk
from nltk.tokenize import sent_tokenize

from tqdm import tqdm
import torch

nltk.download("punkt")
```

```python
# AutoModelForSeq2SeqLM: This is a class from the transformers library that automatically loads a sequence-to-sequence (Seq2Seq) model for tasks like text summarization, translation, and other generation tasks.
# AutoTokenizer: This is a class used to load the appropriate tokenizer associated with the model. A tokenizer converts text into tokens that the model can understand.
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer


device = "cuda" if torch.cuda.is_available() else "cpu"
# AutoTokenizer.from_pretrained(model_ckpt): This line loads the tokenizer associated with the "google/pegasus-cnn_dailymail" checkpoint.
model_ckpt = "google/pegasus-cnn_dailymail"

tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
# AutoModelForSeq2SeqLM.from_pretrained(model_ckpt): This loads the pre-trained Pegasus model from the "google/pegasus-cnn_dailymail" checkpoint. The model is pre-trained and fine-tuned for summarization tasks.
model_pegasus = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt).to(device)
```

```python

def generate_batch_sized_chunks(list_of_elements, batch_size):
    """split the dataset into smaller batches that we can process simultaneously
    Yield successive batch-sized chunks from list_of_elements."""
    for i in range(0, len(list_of_elements), batch_size):
        yield list_of_elements[i : i + batch_size]
```

```python
def calculate_metric_on_test_ds(dataset, metric, model, tokenizer,
                               batch_size=16, device=device,
                               column_text="article",
                               column_summary="highlights"):
    article_batches = list(generate_batch_sized_chunks(dataset[column_text], batch_size))
    target_batches = list(generate_batch_sized_chunks(dataset[column_summary], batch_size))

    for article_batch, target_batch in tqdm(
        zip(article_batches, target_batches), total=len(article_batches)):

        inputs = tokenizer(article_batch, max_length=1024,  truncation=True,
                        padding="max_length", return_tensors="pt")

        summaries = model.generate(input_ids=inputs["input_ids"].to(device),
                         attention_mask=inputs["attention_mask"].to(device),
                         length_penalty=0.8, num_beams=8, max_length=128)
        ''' parameter for length penalty ensures that the model does not generate sequences that are too long. '''

        # Finally, we decode the generated texts,
        # replace the  token, and add the decoded texts with the references to the metric.
        decoded_summaries = [tokenizer.decode(s, skip_special_tokens=True,
                                clean_up_tokenization_spaces=True)
               for s in summaries]

        decoded_summaries = [d.replace("", " ") for d in decoded_summaries]


        metric.add_batch(predictions=decoded_summaries, references=target_batch)

    #  Finally compute and return the ROUGE scores.
    score = metric.compute()
    return score
    # This function evaluates the summarization model on a test dataset by generating summaries for batches of articles and comparing them with the reference summaries using a specified metric (like ROUGE). The function operates in batches to handle large datasets efficiently and returns the calculated metric, which reflects the model's performance on the test data.
```

```python


def convert_examples_to_features(example_batch):
    input_encodings = tokenizer(example_batch['dialogue'] , max_length = 1024, truncation = True )

    with tokenizer.as_target_tokenizer():
        target_encodings = tokenizer(example_batch['summary'], max_length = 128, truncation = True )

    return {
        'input_ids' : input_encodings['input_ids'],
        'attention_mask': input_encodings['attention_mask'],
        'labels': target_encodings['input_ids']
    }
    # The function returns a dictionary containing:
# input_ids: The tokenized dialogue, which will be fed into the model as input.
# attention_mask: The attention mask for the dialogue, helping the model distinguish between actual tokens and padding.
# labels: The tokenized summary, which will be used as the target output during model training.
  #  This code prepares the dataset_samsum dataset for training a text summarization model. The convert_examples_to_features function tokenizes the dialogue and its corresponding summary, creating inputs that the model can use. The dataset is then transformed by mapping this function over all examples, producing a new dataset (dataset_samsum_pt) that is ready for training or evaluation.
dataset_samsum_pt = dataset_samsum.map(convert_examples_to_features, batched = True)
```

```python
from transformers import DataCollatorForSeq2Seq

seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model_pegasus)
```

```python
from transformers import TrainingArguments, Trainer

trainer_args = TrainingArguments(
    output_dir='pegasus-samsum', num_train_epochs=1, warmup_steps=500,
    per_device_train_batch_size=1, per_device_eval_batch_size=1,
    weight_decay=0.01, logging_steps=10,
    evaluation_strategy='steps', eval_steps=500, save_steps=1e6,
    gradient_accumulation_steps=16
)
# If you have a batch of dialogue sequences with varying lengths, the data collator will pad them to the same length, apply the appropriate attention masks, and align the target sequences (summaries) accordingly.
# This code initializes a DataCollatorForSeq2Seq object, which is used to prepare batches of data for a sequence-to-sequence model, such as Pegasus. It handles padding and ensures that the input sequences and their corresponding outputs (labels) are properly aligned, making it easier to train or evaluate the model efficiently.
```

```python


trainer = Trainer(model=model_pegasus, args=trainer_args,
                  tokenizer=tokenizer, data_collator=seq2seq_data_collator,
                  train_dataset=dataset_samsum_pt["train"],
                  eval_dataset=dataset_samsum_pt["validation"])
```

```python

trainer.train()
```

```python

score = calculate_metric_on_test_ds(
    dataset_samsum['test'], rouge_metric, trainer.model, tokenizer, batch_size = 2, column_text = 'dialogue', column_summary= 'summary'
)

rouge_dict = dict((rn, score[rn].mid.fmeasure ) for rn in rouge_names )

pd.DataFrame(rouge_dict, index = [f'pegasus'] )
```

```python
from datasets import load_metric

rouge = load_metric("rouge")

# Your generated summary
summary = summarizer(ARTICLE, max_length=130, min_length=30, do_sample=False)[0]['summary_text']

# Reference summary (replace with actual reference summary)
reference_summary = "Employers are increasingly using biometric time clock machines, such as fingerprint scanners, to track employee attendance. While these devices offer benefits like enhanced security and reduced time theft, they also raise significant privacy and health concerns. The COVID-19 pandemic has heightened worries about the potential spread of the virus through shared surfaces like fingerprint scanners. Additionally, privacy advocates warn that biometric data, which cannot be changed like a password or phone number, poses a risk of identity theft and unwanted surveillance. Some companies have faced legal challenges over their use of these devices, with privacy laws in a few states restricting biometric data collection."

# Compute ROUGE scores
scores = rouge.compute(predictions=[summary], references=[reference_summary])

print(scores)
# ROUGE will give you scores like ROUGE-1, ROUGE-2, and ROUGE-L, which correspond to the overlap of unigrams, bigrams, and the longest common subsequence, respectively.
```

```python
from nltk.translate.bleu_score import sentence_bleu
from nltk.tokenize import word_tokenize

# Tokenize summaries
reference_tokens = word_tokenize(reference_summary)
generated_tokens = word_tokenize(summary)

# Compute BLEU score
bleu_score = sentence_bleu([reference_tokens], generated_tokens)
print(f"BLEU score: {bleu_score}")
# BLEU (Bilingual Evaluation Understudy): Although traditionally used for machine translation, BLEU can also be applied to summarization.
```

```python
from rouge_score import rouge_scorer
import pandas as pd

# Assuming you have a DataFrame with dialogues and their reference summaries
data = {
    'dialogue': [dataset_samsum['test'][i]['dialogue'] for i in range(5)],
    'reference_summary': [dataset_samsum['test'][i]['summary'] for i in range(5)],
    'generated_summary': [summarizer(dialogue, max_length=130, min_length=30, do_sample=False)[0]['summary_text'] for i in range(5)]
}

df = pd.DataFrame(data)

# Initialize ROUGE scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

# Compute ROUGE scores
scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}

for i in range(len(df)):
    reference = df['reference_summary'][i]
    generated = df['generated_summary'][i]
    score = scorer.score(reference, generated)
    for key in scores.keys():
        scores[key].append(score[key].fmeasure)

# Convert scores to DataFrame for easy visualization
scores_df = pd.DataFrame(scores)
print(scores_df.mean())

# ROUGE-1 (Unigram Overlap):
# Ideal Range: 0.3 - 0.5
# ROUGE-2 (Bigram Overlap):
# Ideal Range: 0.1 - 0.3
# ROUGE-L (Longest Common Subsequence):
# Ideal Range: 0.2 - 0.4
```
